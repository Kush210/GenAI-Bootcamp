```python
import time

responses = []
for category, link in links.items():
    time.sleep(20)
    response = requests.get(link, headers={'User-agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36"})
    if response.status_code == 200:
        print(f"Page fetched successfully: {link}")
        responses.append(response)
    else:
        print(f"Failed to fetch page {link}. Status Code: {response.status_code}")
```

```python
categories = {}

for response, category in zip(responses, links.keys()):
    categories[category] = {}
    if response.status_code == 200:
        # Parse the page content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all <a> tags and extract their href attributes
        section = soup.find('section', class_="rsection rsection--with-bg rsection--bg-blue block block-renesas-blocks block-renesas-blocks-application-subcategories", id="block-renesas-blocks-application-subcategories")
        divs = section.find_all('div', class_="rcard rcard--has-gradient")
        for sample in divs:
            sub_category = sample.find('a', class_='rcard__title')["title"]
            print("sub_category", sub_category)
            link = sample.find('a', class_='rcard__title')['href']
            print("link", home+link)

            categories[category][sub_category] = [home+link]
    else:
        print(f"Failed to fetch page. Status Code: {response.status_code}")
categories
```

```python
for category, subcategories in categories.items():
    for subcategory in  subcategories.keys():
        link = categories[category][subcategory][0]
        time.sleep(30)
        response = requests.get(link, headers={'User-agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36"})
        if response.status_code == 200:
            if categories[category][subcategory][1] == None:
                categories[category][subcategory].append({})
            print(f"Page fetched successfully: {link}")
            soup = BeautifulSoup(response.content, 'html.parser')

            # Find all <a> tags and extract their href attributes
            divs = soup.find_all("a", class_="application-category-list__link")

            for sample in divs:
                title = sample["title"]
                print("sub_category", title)
                url = sample['href']
                print("link", home+url)
                categories[category][subcategory][1][title] = [home+url]
            print(categories)

        else:
            print(f"Failed to fetch page: {link}. Status Code: {response.status_code}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Specify the correct path to chromedriver
chrome_driver_path = '/usr/bin/chromedriver'

# Set Chrome options (optional: headless mode for no UI)
chrome_options = Options()
# chrome_options.add_argument("--headless")  # Remove if you want a UI

# Set up the service with the correct path to ChromeDriver
service = Service(chrome_driver_path)

# Create the WebDriver object
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
category = 'Communications Infrastructure'
sub_categories = ['Networking & Fixed Access', 'Wireless Infrastructure']
for subcategory in sub_categories:
    link = categories[category][subcategory][0]
    driver.get(link)
    time.sleep(10)  # Wait for the page to load
    page_source = driver.page_source
    if categories[category][subcategory][1] == None:
        categories[category][subcategory].append({})
    print(f"Page fetched successfully: {link}")
    soup = BeautifulSoup(page_source, 'html.parser')

    # Find all <a> tags and extract their href attributes
    divs = soup.find_all("a", class_="application-category-list__link")

    for sample in divs:
        title = sample["title"]
        print("sub_category", title)
        url = sample['href']
        print("link", home+url)
        categories[category][subcategory][1][title] = [home+url]
    print(categories)
```
