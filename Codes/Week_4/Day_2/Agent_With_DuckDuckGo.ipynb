```python
from crewai import Agent, Task, Crew, Process
from crewai.tools import BaseTool
from pydantic import Field
from langchain_community.tools import DuckDuckGoSearchResults
from crewai.tools import tool
from crewai_tools import SerperDevTool
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()
# Configure environment variables for Groq integration
os.environ["OPENAI_API_BASE"] = "http://localhost:11434/v1"
os.environ["OPENAI_MODEL_NAME"] = "gemma:2b"  
os.environ["OPENAI_API_KEY"] = "NA"

# Ensure your SerperDev API key is set in the environment variables
os.environ["SERPER_API_KEY"] = os.environ["SERPER_DEV_API_KEY"]

@tool("DuckDuckGo Search")
def duckduckgo_search(query: str) -> str:
    """Perform a search using DuckDuckGo."""
    search = DuckDuckGoSearchResults()
    try:
        return search.run(query)
    except Exception as e:
        return f"Error performing DuckDuckGo search: {str(e)}"

serperDev_search = SerperDevTool()

# Choose the search tool to use: 'duckduckgo_search' or 'serperdev_search'
selected_search_tool = duckduckgo_search 

# Create the Research Analyst Agent
researcher = Agent(
    role='Research Analyst',
    goal='Gather current market data and trends',
    backstory="""You are an expert research analyst with years of experience in
    gathering market intelligence. You're known for your ability to find
    relevant and up-to-date market information and present it in a clear,
    actionable format.""",
    tools=[serperDev_search],
    verbose=True
)

# Define the Research Task with the expected_output parameter
research_task = Task(
    description="Research the latest trends in the technology sector.",
    expected_output="A comprehensive report detailing the latest technology market trends.",
    agent=researcher
)

# Create the Crew
crew = Crew(
    agents=[researcher],
    tasks=[research_task],
    process=Process.sequential,
    verbose=True
)

# Run the Crew
crew.kickoff()
```
